```python
"""Analice y trabaje con df2.
Unir los datos resultantes a partir de “df1 y df2” en una sola tabla y exportarlos a Excel. 
El cambio de librerias es aceptable, mientras cumpla el objetivo de la practica. 
Deberá especificar que librerías a utilizado de ser el caso."""
```

```python
documento1 = read_pdf(input_path='EjemploPdf.pdf', pages='2', encoding='windows-1252')
documento1
```

```python
documento2 = read_pdf(input_path='EjemploPdf.pdf', pages='3', encoding = 'windows-1252', pandas_options = {'header': None})
documento2
```

```python
df1 = pd.DataFrame(documento1[0])
df2 = pd.DataFrame(documento2[0])
df2
```

```python
df2.columns = ['State', 'Mar-19', 'Mar-20', 'Growth', '2018-19', '2019-20', 'Growth.1']
df2
```

```python
unir = pd.concat([df1, df2], axis = 0, ignore_index = True)
unir
```

```python
unir.to_excel('tabla.xlsx',index = False)
```
