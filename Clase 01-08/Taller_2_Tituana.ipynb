```python
import pandas as pd
dataFr = pd.read_csv('https://raw.githubusercontent.com/asalber/manual-python/master/datos/colesterol.csv')
dataFr
```

```python
#Renombrar las columnas ‘nombre’ por ‘nombre_apellido’ y ‘altura’ por ‘estatura’.Los index 0,1 y 2 por 1000,1001 y 1003.

```

```python
#Mostrar datos específicos • colesterol 249 • altura 1.94 •Datos index 5 y 7,no mostrar colesterol • Todas las filas,columnas 0,1,5

print(dataFr.loc[5,'colesterol'])

print(dataFr.loc[8,'altura'])
```

```python
# Eliminar filas con datos desconocidos

dataFr.dropna()
```

```python
#Eliminar index 1 y 13.

print(dataFr.drop([1,13]))
```

```python
#Añadir 2 registros al dt.

dataFr = dataFr.append(pd.Series(['Mauricio Carrion','19','H',65.5,1.7,201.3], 
                                 index =['nombre','edad','sexo','peso', 'altura', 'colesterol']), ignore_index=True)

dataFr = dataFr.append(pd.Series(['David Sanchez','20','H',70,1.82,221.5], 
                                 index =['nombre','edad','sexo','peso', 'altura', 'colesterol']), ignore_index=True)

dataFr
```

```python
#Mostrar datos donde Sexo = H y colesterol > 200
print(dataFr[(dataFr['sexo']=='H') & (dataFr['colesterol'] > 200)])
```

```python
#Ordenar los datos de forma descendente acorde al peso.

print(dataFr.sort_values('peso', ascending = False))
```

```python
#Modificar H(hombre) por M y M(mujer) por F

dataFr['sexo'] = ['M' if sn == 'H' else 'F' for sn in dataFr['sexo']]
dataFr
```

```python
#Añadir la columna triglicéridos, para colesterol > 250 True caso contrario False

dataFr['triglicéridos'] = pd.Series(['True' if sn > 250 else 'False' for sn in dataFr['colesterol']])
dataFr
```
