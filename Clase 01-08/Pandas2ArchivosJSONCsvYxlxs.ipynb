# Archivos planos

```python
def crear_arch():
    archivo = open('datos.txt','w') #creo/leo el archivo
    archivo.close() #cierro el archivo
```

```python
def actualizar():
    archivo = open('datos.txt','a') #actualizar 
    archivo.write('Primera Linea 1 \n')
    archivo.write('Segunda Linea 2 \n')
    archivo.write('Tercera Linea 3 \n')
    archivo.close()
```

```python
def leer_arch():
    archivo = open('datos.txt','r') #leer
    linea = archivo.readline()
    while(linea != ""):
        print(linea)
        linea = archivo.readline()
    archivo.close()
```

```python
crear_arch()
```

```python
actualizar()
```

```python
leer_arch()
```

# Ejemplo manipulacion de datos

```python
# Creacion de DataFrame

dataFr = pd.DataFrame({
    'Name':['Ricardo','Isabel','Karla','Fabricio'],
    'Nacimiento':['05-11-2020','20-06-2018','16-08-2017','19-09-2019'],
    'Edad':[43,87,19,33],
    'Peso':[48.5,65.8,57.9,52.56],
    'Sexo':['M','F','F','M'],
})

dataFr
```

```python
#Convertir fecha de String a Date
pd.to_datetime(dataFr.Nacimiento, format = '%d-%m-%Y')
dataFr
```

```python
#Eliminar columnas
edad = dataFr.pop('Edad')
edad
```

```python
dataFr
```

```python
#AÃ±air filas
dataFr = dataFr.append(pd.Series(['Mauricio Carrion','13-12-2000',61.35,'M'],
                        index =['Name','Nacimiento','Peso','Sexo']), ignore_index = True)

dataFr.tail() #
```

```python
#Eliminar filas
print(dataFr.drop([5,7]))
```

```python
dataFr = dataFr.drop([5,7])
dataFr
```

```python
#Filtrar datos
print(dataFr[(dataFr['Sexo']=='F') & (dataFr['Peso'] > 57)])
```

```python
#Ordenar tomando en cuenta las columnas

print(dataFr.sort_values('Peso'))
```

```python
#AÃ±adir columnas al DataFrame

dataFr['Nuevo'] = pd.Series([True,False,True,False,True,False])
dataFr
```

```python
dataFr['Nuevo'] = ['True' if sn >= 57 else 'false' for sn in dataFr['Peso']]
dataFr
```

```python
#Eliminar dtos desconocidos

dataFr.dropna()
```
