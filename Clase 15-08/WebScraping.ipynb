```python
import requests
from bs4 import BeautifulSoup
```

```python
pagina_web="https://subslikescript.com/movie/Titanic-120338"
result = requests.get(pagina_web)
contenido = result.text
soup = BeautifulSoup(contenido,'lxml')
soup
```

```python
resumen = soup.find('article', class_='main-article')
resumen
```

```python
titulo = soup.find('h1').get_text()
titulo
```

```python
contenido2 = soup.find('div', class_='full-script').get_text()
contenido2
```

```python
from unidecode import unidecode
with open(f'{titulo}.txt', 'w') as file:
    file.write(unidecode(contenido2))
```

# Etiqueta y posicion de elementos HTML

```python
consulta = requests.get('https://news.ycombinator.com')
soup = BeautifulSoup(consulta.text,'html.parser')
buscar = soup.findAll('tr', class_='athing')
```

```python
lista = list()

for datos in buscar:
    data = {
        'id': datos['id'],
        'title': datos.find_all('td')[2].a.text,
        'rank': int(datos.find_all('td')[0].span.text.replace('.','')),
        'url': datos.find_all('td')[2].a['href']
    }
    lista.append(data)
    
```

```python
df = pd.DataFrame(lista)
df
```

```python
import csv
filename = 'consulta.csv'
with open(filename,'w',newline='') as f:
    w = csv.DictWriter(f,['id','title','rank','url'])
    w.writeheader()
    for data in lista:
        w.writerow(data)
```

# Guardar Base de Datos

```python
import sqlite3
```

```python
con = sqlite3.connect("BDD_DatosWeb.db")

try:
    con.execute("""CREATE TABLE information (
                    id integer,
                    title text,
                    rank integer,
                    url text
                )""")
    print("Tabla creada")
except sqlite3.OperationalError:
    print("La tabla ya existe")
    
```

```python
datos = requests.get('https://news.ycombinator.com')
soup2 = BeautifulSoup(datos.text,'html.parser')
busqueda = soup2.findAll('tr', class_='athing')
cur = con.cursor()

for datos in busqueda:
    cur.execute("INSERT INTO information (id,title,rank,url) VALUES(?,?,?,?)",(
                    datos['id'],
                    datos.find_all('td')[2].a.text,
                    int(datos.find_all('td')[0].span.text.replace('.','')),
                    datos.find_all('td')[2].a['href']
                ))
    
con.commit()
cur.close()
con.close()
```
