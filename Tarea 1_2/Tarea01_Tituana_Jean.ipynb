```python
fileOrigin=open('Starwars.txt')

file=fileOrigin.read()
files=file.split()

fileOrigin.close()

files
```

```python
"""1. Guarde el archivo resultante (file2) como sin_duplicados.txt"""

duplicados = open('sin_duplicados.txt','w')

file2 = []

for i in files:
    if i not in file2:
        file2.append(i)

for item in file2:
    print(f'{item} \n')
    duplicados.write("%s\n" %item)

duplicados.close()
```

```python
"""2. Contar el número de ocurrencias de las palabras de un archivo de texto. Ejm:.
• Brackett , 1
• from , 127
• a , 538
Guardar el resultado en un archivo .txt"""


dic = {}

archivo = open('ocurrencias.txt','w')

for w in files:
    dic[w] = files.count(w)
    
for (l,n) in dic.items():
    print(f'{l} -> {n} \n')
    archivo.write(f'{l} -> {n} \n')
    
archivo.close()
```

```python
"""3. Contar el número de ocurrencias donde las palabras tengan más de 4 caracteres. Ejm:.
• destroyed, ,1
• Imperial , 42
• troops , 25
Guardar el resultado en un archivo .txt"""

dic2 = {}

for w in files:
    if(len(w)>4):
        dic2[w] = files.count(w)
    
archivo2 = open('mayores_a_cuatro.txt','w')
for (l,n) in dic2.items():
    print(f'{l} -> {n} \n')
    archivo2.write(f'{l} -> {n} \n')
archivo2.close()
```
